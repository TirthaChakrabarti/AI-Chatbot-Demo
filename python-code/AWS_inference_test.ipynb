```python
import boto3
import json

client = boto3.client(
    service_name='bedrock-runtime',
    region_name='us-east-1'
)

model_inference_profile = 'arn:aws:bedrock:us-east-1:823413233438:inference-profile/us.meta.llama3-1-8b-instruct-v1:0'
```

```python
def get_chatbot_response(client, model_name, messages, temperature=0.0):
    # Prepare prompt
    # input_messages = []
    # for message in messages: 
    #     input_messages.append({"role": message["role"], "content": message["content"]})

    # Send request to AWS Bedrock
    response = client.invoke_model(
        modelId=model_name,
        body=json.dumps({
            "max_gen_len": 2000,
            "temperature": temperature,
            "top_p": 0.8,
            "prompt": "[INST] " + messages + " [/INST]"
        })
    )

    # Parse JSON response
    response_body = json.loads(response['body'].read())
    return response_body.get("generation", "").strip()
```

```python
# messages = [
#     {"role": "user", "content": "Highest mountain peak of Earth?"}
# ]

messages = "Highest mountain peak of Earth? Single word answer."

response = get_chatbot_response(client, model_inference_profile, messages)

print(response)
```
